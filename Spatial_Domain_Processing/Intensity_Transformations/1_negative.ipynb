### Image Negatives

An **image negative** is obtained by applying a **negative transformation** to a digital image, where the intensity levels of each pixel are reversed. Mathematically, if the original intensity levels range from 0 to \(L-1\), each pixel value \(r\) is transformed to:

```
s = L - 1 - r
```

where:

* `r`: input intensity value  
* `s`: output intensity value  
* `L`: the maximum intensity level (e.g., 256 for an 8-bit image, where `n = 8` bits)

This transformation effectively **inverts the intensity histogram** of the image, producing the equivalent of a photographic negative. Bright regions become dark, and dark regions become bright. 

#### Detailed Explanation

The negative transformation is a **point operation**, meaning it maps each input pixel independently to an output pixel without considering its neighbors. The transformation function is linear and monotonic decreasing, which preserves relative contrast but reverses the polarity of the image intensities.

One key property is that areas with **low intensity values** (dark regions) are mapped to **high intensity values** in the negative image, making subtle details more visible. This is particularly useful in images where dark regions dominate, as it enhances structures that might otherwise be overlooked.

#### Applications

1. **Medical Imaging:**  
   Negative images are commonly used in radiography and mammography. For instance, in digital mammograms, small lesions or microcalcifications in dense tissue may be easier to detect in the negative image.

2. **Astronomical Imaging:**  
   Astronomers often use image negatives to detect faint stars or galaxies in dark regions of space.

3. **Document Analysis:**  
   In digitized historical documents, negative images can help in enhancing faded ink or handwritten text.

4. **Preprocessing for Edge Detection:**  
   In some computer vision pipelines, applying a negative transformation can improve the performance of edge detection algorithms in images with predominantly dark backgrounds.

#### Summary

While the visual content of the original and negative images remains the same, the negative transformation emphasizes structures that are visually subtle in dark regions. It is a simple yet powerful technique in **image enhancement** and forms the basis for many other **intensity transformation methods** in digital image processing.


````python
import cv2
import matplotlib.pyplot as plt
# cv2.IMREAD_GRAYSCALE --> to convert it Momentarily to GRAYSCALE
image = cv2.imread('Images/1_0_image.png', cv2.IMREAD_GRAYSCALE) 

# number of bits = 8, thus L - 1 = (2^n) - 1 ==> L - 1 = (2^8) - 1 = 255  
negative_image = 255 - image

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image, cmap="gray")
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(negative_image, cmap="gray")
plt.title('Negative Image')
plt.axis('off')

plt.show()
````
